```python
# Calculate the gender ,value of counts?
d[['Gender']].value_counts()
```

```python
a=sns.countplot(d,x='Gender')
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
#Analyze selling amount percentage by gender
a=d.groupby('Gender')['Amount'].sum()
total=a.sum()
percentage=a/total*100
```

```python
plt.bar(percentage.index,percentage.values,color=['brown','green'])
plt.title('Selling Amount Percentege By Gender')
plt.xlabel('Gender')
plt.ylabel('Amount in percentage')
plt.show()
```

```python
#Display count of AGe Group?
d['Age Group'].value_counts()
```

```python
sns.countplot(d,x='Age Group')
plt.show()
```

```python
#Display count of marital-status?
d['Marital_Status'].value_counts()
```

```python
sns.countplot(d,x='Marital_Status')
plt.show()
```

```python
#There is a ratio of Marital_status in using a pie chart?
plt.pie(list(d['Marital_Status'].value_counts()),labels=list(d['Marital_Status'].value_counts().keys()),autopct='%0.1f%%',startangle=90,
       counterclock=False)
plt.show()
```

```python
#How many product_category in our data?
d['Product_Category'].value_counts()
```

```python
#How many ration of product_category?Using pie chart.
plt.figure(figsize=(10,30))
d['Product_Category'].value_counts().plot(kind="bar")
plt.ylabel('Number')
plt.show()
```

```python
#How many total gender should I purchase the orders?
a=d.groupby('Gender')['Orders'].sum()
a
```

```python
plt.bar(a.index,a.values,color=['Red','Yellow'])
plt.title('Sum of total order between gender')
plt.xlabel('Gender')
plt.ylabel('No. of Orders')
plt.show()
```

```python
#Analyze sale of age_group?
z=d.groupby('Age Group')['Amount'].sum()
z
```

```python
sns.barplot(x=z.index,y=z.values,data=z.reset_index())
plt.title('Age group between Amount')
plt.show()
```

```python
#How many sales in every state?
n=d.groupby('State')['Amount'].sum().nlargest(10)
n
```

```python

plt.figure(figsize=(14,8))
sns.barplot(x=n.values, y=n.index, data=n.reset_index(),orient='h',hue='State',palette='viridis')
plt.title('Total Sales by State')
plt.xlabel('Amount')
plt.ylabel('State')
plt.show()
```

```python
#Calculate the orders in Every state?
a=d.groupby('State')['Orders'].sum()
a
```

```python
plt.bar(x=a.index,height=a.values,data=a)
plt.title("Sum of orders in every state")
plt.xlabel('State')
plt.ylabel('Sum of orders')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze total sales by marital status and gender
a=d.groupby(['Marital_Status','Gender'])['Amount'].sum()
a
```

```python
sns.barplot(x='Marital_Status',y='Amount', hue='Gender',data=a.reset_index(),palette='Set2')
plt.title('Total Sales of marital_status')
plt.show()
```

```python
#Calculate the Occupation?
b=d['Occupation'].value_counts()
b
```

```python
sns.countplot(x='Occupation',data=d)
plt.title('Counts of Occupation')
plt.xticks(rotation=90)
plt.show()
```

```python
t=d.groupby('Occupation')['Amount'].sum()
t
```

```python
sns.barplot(x=t.values,y=t.index,data=t.reset_index(),hue='Occupation',orient='h',palette='rocket')
plt.show()
```
